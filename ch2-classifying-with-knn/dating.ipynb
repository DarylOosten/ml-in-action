```python
labels = ['frequent_flyer_miles', 'percentage_of_time_playing_video_games', 'liters_of_icecream_per_week', 'like']
attributes = ['frequent_flyer_miles', 'percentage_of_time_playing_video_games', 'liters_of_icecream_per_week']
sample_size = 100
k = 5
```

```python
data = pd.read_csv('datingTestSet.txt', delimiter='\t', names=list(labels))
data.head()
```

```python
r = np.random.choice(data.index, sample_size, replace=False)
sample = data.ix[r]
sample.describe()
```

```python
data = data.drop(r)
data.describe()
```

```python
sample_like = sample['like']
sample_like.describe()
```

```python
data_like = data['like']
data_like.describe()
```

```python
sample = sample[attributes]
data = data[attributes]
```

```python
normalized_data = (data - data.min()) / (data.max() - data.min())
normalized_data.describe()
```

```python
normalized_sample = (sample - data.min()) / (data.max() - data.min())
normalized_sample.describe()
```

```python
euclidean = lambda candidate: lambda value: abs(candidate - value) ** 2

def classify(candidate):
    scores = normalized_data.apply(euclidean(candidate), axis=1).sum(axis = 1)
    scores = pd.concat([scores, data_like], axis = 1, keys = list(['score', 'like']))
    scores = scores.sort(columns=['score'])
    scores = scores.head(k)['like']
    return scores.value_counts().keys()[0]

classified = normalized_sample.apply(classify, axis = 1)
classified = pd.concat([classified, sample_like], axis = 1, keys = list(['should_like', 'does_like']))
classified.head()
```

```python
hits = classified.apply(lambda r: r[0] == r[1], axis = 1)
hits = pd.DataFrame(hits.value_counts()).T
hits.columns = ['hits', 'misses']
hits['sample_size'] = sample_size
hits
```
